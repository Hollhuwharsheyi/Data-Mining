```python
import scrapy
from scrapy.crawler import CrawlerRunner
from crochet import setup, wait_for
setup()
```

```python
class PubSpdr(scrapy.Spider):
  name = "Sefas"

  start_urls = [
    "https://pureportal.coventry.ac.uk/en/organisations/school-of-economics-finance-and-accounting/publications/"
  ]

  custom_settings = {
      "FEEDS": {
          "articles.csv": {
              "format": "csv",
              "overwrite": True
          }
      }
  }



  def parse(self, response):
    pages = response
    li=pages.css('li.list-result-item')
    title = li.css('h3.title span::text').get()
    for lib in pages.css('li.list-result-item'):
      tit = lib.css('h3.title span::text').get()
      auth = lib.css('a.person span::text').getall()
      pub_dt = lib.css('span.date::text').get()
      pub_lnk = lib.css ('h3.title a.link::attr(href)').extract()
      auth_lnk = lib.css('a.person::attr(href)').getall()
      yield{
          "Authors":auth,
          "Title":tit,
          "Author Profile":auth_lnk,
          "Publication Link":pub_lnk,
          "Date":pub_dt
      }
    
    nxt_pages = response.css('a.nextLink::attr(href)').extract()
    if nxt_pages is not None:
      nxt_pages = response.urljoin(nxt_pages[0])
      print(nxt_pages)
      yield scrapy.Request(nxt_pages, callback=self.parse)
    
  custom_settings = {
      "FEEDS": {
          "articles.csv": {
              "format": "csv",
              "overwrite": True
          }
      }
  }

@wait_for(30)
def run_spider():
  my_spider = CrawlerRunner()
  spider = my_spider.crawl(PubSpdr)
  return spider

run_spider()
```

```python
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
nltk.download('punkt')
nltk.download("stopwords")
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
#df = pd.read_csv('/content/articles.csv')
file_path = "/content/articles.csv"
def inverted_index(file_path):
  df = pd.read_csv(file_path)
  df = df['Title']
  Inverted_keytable={}

  doc_id = 0
  for i in df:
    #print("title -", doc_id)
    tokens = word_tokenize(i)
    tokens_lower = [w.lower() for w in tokens if w.isalnum()]
    re_stopwords = [w for w in tokens_lower if w not in stopwords.words('english')]
    stemmed = [stemmer.stem(w) for w in re_stopwords]

    for stem in stemmed:
      value = Inverted_keytable.get(stem)
      if value == None:
        l = [1, [doc_id]]
        Inverted_keytable[stem] = l
      else:
        l = Inverted_keytable[stem]
        #list[o] = df[stem]
        #list[1] = pl[stem]
        if doc_id not in l[1]:
          l[1].append(doc_id)
          l[0] += 1

    doc_id += 1

  return Inverted_keytable
```
